# dictionary to list

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
merge['genres'] = merge['genres'].apply(convert)
merge.head()
```

```python
merge.iloc[0].keywords
```

```python
merge['keywords'] = merge['keywords'].apply(convert)
merge.head()
```

```python
merge.iloc[0].cast
```

```python
merge['cast'] = merge['cast'].apply(convert)
merge.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
merge['cast'] = merge['cast'].apply(collapse)
merge['genres'] = merge['genres'].apply(collapse)
merge['keywords'] = merge['keywords'].apply(collapse)
merge.head()
```

```python
merge.head()
```
